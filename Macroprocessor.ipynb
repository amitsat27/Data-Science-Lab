# MACRO CODE WITHOUT PARAMETERS

```python
import re
from prettytable import PrettyTable
tokens = list()
code = open("MACCODE1.txt",encoding="utf-8")
line = code.readlines()
for i in line:
    k = i.strip('')
    k = i.strip('\n')
    k = re.split(' |,',k)
    tokens.append(k)
    
for i in tokens:
    for j in i:
        if(j==''):
            i.remove(j)
final_tokens = [x for x in tokens if x!=[]]
final_tokens
flag = 0

macrodict = dict()
IC_mdt,IC_mnt,mntp = 1,1,1
macros = 0


mnt = PrettyTable(['Name of Macro','No. of Parameters','Starting Index'])
mdt = PrettyTable(['Index','Body'])
expandedcode = PrettyTable(['Expanded Code'])
Intermediate = PrettyTable(['Intermediate Code'])


for i in range(len(final_tokens)):
    if(final_tokens[i][0]=='MACRO' and len(final_tokens[i])==2):
        macrotrack = list()
        mnt.add_row([final_tokens[i][1],0,IC_mnt])
        macrodict[final_tokens[i][1]]= ''
        IC_mnt = IC_mnt + 1
        macros = i
        flag = 1
    elif(flag == 1 and final_tokens[i][0]!='MEND'):
        if(len(final_tokens[i]) <3):
            mdt.add_row([IC_mdt,final_tokens[i][0]+" "+final_tokens[i][1]])
            macrotrack.append(final_tokens[i][0]+" "+final_tokens[i][1])
        else:
            mdt.add_row([IC_mdt,final_tokens[i][0]+" "+final_tokens[i][1]+","+final_tokens[i][2]])
        IC_mnt = IC_mnt+1
        IC_mdt = IC_mdt+1
    elif(final_tokens[i][0]=='MEND'):
        mdt.add_row([IC_mdt,final_tokens[i][0]])
        IC_mdt = IC_mdt+1
        macrodict[final_tokens[macros][1]]=macrotrack
        flag = 0
    elif(flag==0):
        if final_tokens[i][0] not in macrodict.keys():
            if(len(final_tokens[i]) == 1 ):
                Intermediate.add_row([final_tokens[i][0]])
                expandedcode.add_row([final_tokens[i][0]])
            elif(len(final_tokens[i]) == 2 ):
                Intermediate.add_row([final_tokens[i][0]+" "+final_tokens[i][1]])
                expandedcode.add_row([final_tokens[i][0]+" "+final_tokens[i][1]])
            else:
                s=""
                for p in range(1,len(final_tokens[i])):
                    s = s + final_tokens[i][p]+","
                Intermediate.add_row([s])
                expandedcode.add_row([s])
        else:
            Intermediate.add_row([final_tokens[i][0]])
            for n in macrodict[final_tokens[i][0]]:
                expandedcode.add_row([n])
            
print(macrodict)
print("MACRO NAME TABLE(MNT)")            
print(mnt)
print("MACRO DEFINITION TABLE(MDT)")
print(mdt)
print("INTERMEDIATE TABLE")
print(Intermediate)
print(expandedcode)
```

```python
import re
from prettytable import PrettyTable
tokens = list()
code = open("MACCODE2.txt",encoding="utf-8")
line = code.readlines()
for i in line:
    k = i.strip('')
    k = i.strip('\n')
    k = re.split(' |,',k)
    tokens.append(k)
    
for i in tokens:
    for j in i:
        if(j==''):
            i.remove(j)
final_tokens = [x for x in tokens if x!=[]]
final_tokens
flag,flag_param = 0,0

macrodict = dict()
mcount = 0
macroslist = list()
IC_mdt,IC_mnt,mntp = 1,1,1
macros = 0
parametertraver,lp=0,0


mnt = PrettyTable(['Name of Macro','No. of Parameters','Starting Index'])
mdt = PrettyTable(['Index','Body'])
expandedcode = PrettyTable(['Expanded Code'])
Intermediate = PrettyTable(['Intermediate Code'])

formal_actual = dict()


for i in range(len(final_tokens)):
    if(final_tokens[i][0]=='MACRO' ):
        macrotrack = list()
        no_of_parameters = len(final_tokens[i]) - 2
        if(no_of_parameters > 0):
            macroslist.append(final_tokens[i][1])
            counter = 1
            parameters = list()
            flag_param = 1
            actual_param = dict()
            templist = dict()
            
            for par in range(2,len(final_tokens[i])):
                final = list()
                t = list(final_tokens[i][par])
                final.append("#"+str(counter))
                templist[t[1]]= final
                counter = counter +1
            parameters.append(templist)
            formal_actual[final_tokens[i][1]] = parameters
        mnt.add_row([final_tokens[i][1],no_of_parameters,IC_mnt])
        
        macrodict[final_tokens[i][1]]= ''
        IC_mnt = IC_mnt + 1
        macros = i
        flag = 1
    elif(flag == 1 and final_tokens[i][0]!='MEND'):
        if(flag_param == 1 and len(final_tokens[i][1]) > 1):
            y = formal_actual[macroslist[lp]]
            mdt.add_row([IC_mdt,final_tokens[i][0]+" "+str(y[0][final_tokens[i][1][1]][0])])
            macrotrack.append(final_tokens[i][0]+" "+final_tokens[i][1])
        else:
            mdt.add_row([IC_mdt,final_tokens[i][0]+" "+final_tokens[i][1]])
            macrotrack.append(final_tokens[i][0]+" "+final_tokens[i][1])
        IC_mnt = IC_mnt+1
        IC_mdt = IC_mdt+1
       
    elif(final_tokens[i][0]=='MEND'):
        if(flag_param == 1):
            lp = lp+1
        flag_param=0
        mdt.add_row([IC_mdt,final_tokens[i][0]])
        IC_mdt = IC_mdt+1
        macrodict[final_tokens[macros][1]]=macrotrack
        flag = 0
        
    elif(flag==0):
        if final_tokens[i][0] not in macrodict.keys():
            if(len(final_tokens[i]) == 1 ):
                Intermediate.add_row([final_tokens[i][0]])
                expandedcode.add_row([final_tokens[i][0]])
            elif(len(final_tokens[i]) > 1 ):
                s=""
                for p in range(len(final_tokens[i])):
                    s = s + final_tokens[i][p]+" "
                Intermediate.add_row([s])
                expandedcode.add_row([s])
        else:
            s=""
            actuallist = list()
          
            if(len(final_tokens[i])>1):
                for num in range(1,len(final_tokens[i])):
                    actuallist.append(final_tokens[i][num])
            for p in range(len(final_tokens[i])):
                s = s + final_tokens[i][p]+" "
            Intermediate.add_row([s])
            c = 0
            for n in macrodict[final_tokens[i][0]]:
                if(len(n.split(" ")[1])>1):
                    formal_actual[final_tokens[i][0]][0][n.split(' ')[1][1]].append(actuallist[c])
                    c=c+1
            for n in macrodict[final_tokens[i][0]]:
                if(len(n.split(" ")[1])>1):
                    d = formal_actual[final_tokens[i][0]][0][n.split(' ')[1][1]][1]
                    finaltext = n.replace(n.split(' ')[1],d)
                    expandedcode.add_row([finaltext])
                else:
                    expandedcode.add_row([n])
                    

print(formal_actual)
for lists in formal_actual.keys():
    formal_table = PrettyTable(['Formal Parameter','Positional Parameter'])
    actual_table = PrettyTable(['Actual Parameter','Positional Parameter'])
    for keys,values in formal_actual[lists][0].items():
        formal_table.add_row([keys,values[0]])
        actual_table.add_row([values[1],values[0]])
    print(formal_table)
    print(actual_table)
print("MACRO NAME TABLE(MNT)")            
print(mnt)
print("MACRO DEFINITION TABLE(MDT)")
print(mdt)
print("INTERMEDIATE TABLE")
print(Intermediate)
print(expandedcode)
```
